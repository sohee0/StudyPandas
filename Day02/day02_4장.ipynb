```python
df1 = df[df['dataset']=='I']
df2 = df[df['dataset']=='II']
df3 = df[df['dataset']=='III']
df4 = df[df['dataset']=='IV']
```

```python
df1
```

```python
fig = plt.figure(figsize=(12,8))
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
axes1.plot(df1['x'],df1['y'],'mo')
axes2.plot(df2['x'],df2['y'],'r+')
axes3.plot(df3['x'],df3['y'],'g>')
axes4.plot(df4['x'],df4['y'],'c^')
axes1.set_title('part1')
fig.suptitle('anscombe')
fig.tight_layout()
```
